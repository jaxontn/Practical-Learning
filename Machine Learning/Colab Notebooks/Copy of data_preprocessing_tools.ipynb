# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
#we need feature variable(s) & dependant variable (last column)
#x means feature variable
#y means dependant variable
x = dataset.iloc[:, :-1].values #using a pandas dataframe   .iloc[row, columns]
y = dataset.iloc[:, -1].values
```

```
print(x)
```

```
print(y)
```

## Taking care of missing data

```
#method 1
#replacing the missing data by the average of the whole data in current column

from sklearn.impute import SimpleImputer #getting the class from sci-kit learn
imputer =  SimpleImputer(missing_values = np.nan, strategy = 'mean') # we want to replcae the missing value
#imputer is an object
imputer.fit(x[:, 1:3]) #for real nubmers
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

```
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
#one hut coding
#replacing string names with numbers

#using 2 classes
#column transform class, the compose module of the sci-kit learn library
#one hot encoder class, the proprocessing module of the same sci-kit learn library

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
#ct means column transformer
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough') 
#[0] for index zero (country coloumn)
x = np.array(ct.fit_transform(x))
```

```
print(x)
```

### Encoding the Dependent Variable

```
#want to change 'No'to 0, 'Yes' to 1
#algorithm for binary outcome

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y) # no need numpy array because it is dependent variable
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
#using sci-kit learn again

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1) 
#recommended size of the split
#80% observation in the training set
#20% observation in the test set
```

```
print(x_test)
```

```
print(y_test)
```
