```python
%matplotlib inline
import pandas as pd
import mplfinance as mpf
```

```python
file = '/Users/USER/Desktop/mplfinance/XAUUSD.csv'
intraday = pd.read_csv(file)
```

```python
intraday.dates = pd.to_datetime(intraday.dates) #to change the data type of 'Date' to datetime64
intraday.info()
```

```python
intraday = intraday.set_index('dates')
```

```python
intraday
```

```python
iday = intraday.loc['2021-04-12 19:45' : '2021-04-12 23:59', :]
mpf.plot(iday, type='candle', volume=True)
```

```python
mpf.plot(iday, type='line', volume=True)
```

```python
mpf.plot(iday['2021-04-12 19:45' : '2021-04-12 21:45'], type='candle', volume=True)
```

```python
mpf.plot(iday['2021-04-12 20:00' : '2021-04-12 20:15'], type='candle', volume=True)
```

```python
mpf.plot(iday['2021-04-12 20:00' : '2021-04-12 23:15'], type='candle', mav=(20), volume=True)
```

```python
mpf.plot(iday['2021-04-12 20:00' : '2021-04-12 23:15'], type='candle', mav=(20), tight_layout=True, volume=True)
```

```python
mpf.plot(iday['2021-04-12 20:00' : '2021-04-12 23:15'], figratio=(20,12), 
         type='candle',
         title = 'XAUUSD Price 2021',
         mav=(20), volume=True, 
         tight_layout=True)
```

```python
mpf.plot(iday['2021-04-12 20:00' : '2021-04-12 23:15'], figratio=(20,12), 
         type='candle',
         title = 'XAUUSD Price 2021',
         mav=(20), volume=True,
         tight_layout=True,
         style = 'charles')
```
