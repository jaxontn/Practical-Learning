```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation
```

```python
plt.rcParams['animation.html'] = 'jshtml'
```

```python
def projectile(V,t,alpha,g=9.81):
  '''
  This is a function to calculate a projectile motion
  '''
  x = V * t * np.cos(alpha)
  y = V * t * np.sin(alpha) - (1/2) * g * t ** 2

  return x, y
```

```python
time = np.linspace(0,20,10)
def animate_func(frames):
  my_x, my_y = projectile(100,time[frames], np.pi/4)

  plt.cla()
  plt.scatter(my_x, my_y)
  plt.xlim(0,1200)
  plt.ylim(0,500)
  plt.title('Time = ' + str(np.round(time[frames],2)) + ' sec')

fig, ax = plt.subplots()
matplotlib.animation.FuncAnimation(fig, animate_func, frames=10)
```
