```python
V = 100 #velocity, unit is m/s
alpha = np.pi / 4 #radian equivilent to 45 degree
g = 9.81  #m/s^2
t = np.linspace(0,20,50)  #sec
x = V * t * np.cos(alpha)
y = V * t * np.sin(alpha) - (1/2) * g * t ** 2
```

```python
plt.scatter(x,y, c=t)
plt.plot([0,1400], [0,0], 'r')
plt.xlabel('Distance x (m)')
plt.ylabel('Distance y (m)')
plt.xlim(0,1200)
plt.ylim(0,300)
plt.colorbar(label='Time (sec)')
```

```python
def projectile(V,t,alpha,g=9.81):
  x = V * t * np.cos(alpha)
  y = V * t * np.sin(alpha) - (1/2) * g * t ** 2

  return x, y
```

```python
for gravity in [2,4,6,8,10]:
  x, y = projectile(100,np.linspace(0,20,50), np.pi/4, gravity)
  plt.scatter(x,y, label=gravity)
  plt.legend(title='Gravity')
  plt.grid()
```
